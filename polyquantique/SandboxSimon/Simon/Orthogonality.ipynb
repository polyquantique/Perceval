```python
import numpy as np
```

```python
def overlap(x1,x2,g=1):
    a = np.exp(-g/2*np.abs(x1-x2))
    return a
```

```python
N=100
x=np.random.standard_cauchy(N)*0.1
print(x)
```

```python
O = np.zeros((N,N))
for i in range(N):
    O[i,i]=1
    for j in range(i):
        O[i,j]=overlap(x[i],x[j])
        O[j,i]=O[i,j]

print(O)

```

```python
L,U=np.linalg.eigh(O)
    
```

```python
print(L)
```

```python
import matplotlib.pyplot as plt
plt.semilogy(L)
```

```python
print(U[-1,:])
```

```python
z=np.linspace(-13,13,100)
E = np.zeros((N,100))
for k in range(N):
    E[k,:]=np.exp(-(z-x[k])/2)*np.heaviside(z-x[k],0)

plt.plot(E.T)
```

```python

for j in range(5):
    eig=np.zeros(N)
    for i in range(N):
        eig = eig + U[-j-1,i] * E[i]
    plt.plot(eig)


plt.show()
```
